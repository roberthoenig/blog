```python
train_csv = pd.read_csv(path.join(datadir, 'train.csv'))
test_csv = pd.read_csv(path.join(datadir, 'test.csv'))
```

```python
test_data = test_csv[['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked', 'PassengerId']]
training_data = train_csv[['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked', 'PassengerId']]
```

```python
training_data = training_data[training_data.apply(lambda row: not row.isnull().values.any(), axis=1)]
test_data = test_data[test_data.apply(lambda row: not row.isnull().values.any(), axis=1)]
```

```python
all_data = pd.concat([training_data, test_data])
attributes = all_data.columns
attribute_values = {attribute: all_data[attribute].unique() for attribute in attributes}
```

```python
def boolean_entropy(probability):
    return scipy.stats.entropy([probability, 1-probability], base=2)
```

```python
class DecisionTreeNode:
    def __init__(self, examples, attributes, parent_examples=None):
        self.children = {}
        self.decision = None
        self.attribute = None
        if len(examples) == 0:
            self.decision = parent_examples['Survived'].value_counts().idxmax()
        elif len(attributes) == 0:
            self.decision = examples['Survived'].value_counts().idxmax()
        elif len(examples['Survived'].unique()) == 1:
            self.decision = examples['Survived'].unique()[0]
        else:
            def get_attribute_entropy(attribute):
                entropy = 0
                for value in attribute_values[attribute]:
                    subset = examples[examples[attribute]==value]
                    survivor_cnt = subset['Survived'].value_counts().get(1, 0)
                    if len(subset) != 0:
                        entropy += (len(subset)/len(examples)) * boolean_entropy(survivor_cnt/len(subset))
                return entropy
            
            self.attribute = min(attributes, key=get_attribute_entropy)
            new_attributes = [a for a in attributes if a != self.attribute]
            for value in attribute_values[self.attribute]:
                subset = examples[examples[self.attribute]==value]
                self.children[value] = DecisionTreeNode(subset, new_attributes, examples)
                    
    def classify(self, example):
        if self.decision is not None:
            return self.decision
        else:
            attribute_value = example[self.attribute]
            return self.children[attribute_value].classify(example)
        
    def tographviz(self, _graph=pgv.AGraph(), _ctr=0):
        _ctr += 1
        nodename = _ctr
        if self.decision == None:
            _graph.add_node(nodename, label=self.attribute)
            for key, val in self.children.items():
                graph, rootname, newctr = val.tographviz(_graph, _ctr)
                _ctr = newctr
                graph.add_edge(nodename, rootname, label=key)
        else:
            _graph.add_node(nodename, label=str(bool(self.decision)))
        return (_graph, nodename, _ctr) 
```

```python
my_training_data = training_data[:700]
my_test_data = training_data[701:]
```

## Local test

```python
decision_attributes = [attr for attr in attributes if attr not in ('Survived', 'PassengerId')]
decision_tree = DecisionTreeNode(my_training_data, decision_attributes)
```

```python
predicted = []
expected = []
for idx, example in my_test_data.iterrows():
    decision = decision_tree.classify(example)
    predicted.append(bool(decision))
    expected.append(bool(example['Survived']))
```

```python
correct_predictions = list(map(lambda tpl: tpl[0]==tpl[1], zip(predicted, expected))).count(True)
all_predictions = len(predicted)
print("Accuracy: {:.0%}".format(correct_predictions/all_predictions))
```

```python
graph, _, _ = decision_tree.tographviz()
imagepath = path.join(datadir, 'decisiontree.png')
graph.layout(prog='dot')
graph.draw(imagepath)
Image(filename=imagepath)
```

## Kaggle test

```python
decision_attributes = [attr for attr in attributes if attr not in ('Survived', 'PassengerId')]
decision_tree = DecisionTreeNode(training_data, decision_attributes)
```

```python
submission = {'PassengerId': [], 'Survived': []}
for idx, example in test_data.iterrows():
    survived = decision_tree.classify(example)
    passenger_id = example['PassengerId']
    submission['PassengerId'].append(passenger_id)
    submission['Survived'].append(survived)
```

```python
submission_file = path.join(datadir, 'submission.csv')
df = pd.DataFrame(submission)
df.to_csv(submission_file, index=False)
```
